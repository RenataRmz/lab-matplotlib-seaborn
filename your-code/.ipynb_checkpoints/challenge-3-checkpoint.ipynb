#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here
titanic_cor = titanic.corr()
titanic_cor
```

```python
#triu_indices_from regresa los valores del triangulo superior de la matriz
mask = np.zeros_like(titanic_cor,dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

#asignar paleta de colores (colores HUSL)
cmap = sns.diverging_palette(6,255, as_cmap =  True)

sns.heatmap(titanic_cor,mask = mask, cmap = cmap, linewidths=0.5,annot= True)
plt.show()
```
